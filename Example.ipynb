```python
import numpy as np
from pylab import *
z = np.linspace(0,2048,2048)
x = np.sin(50*np.pi*z)
y = np.cos(50*np.pi*z)
x2 = 2*np.sin(50*np.pi*z)

plot(x, 'k--')
plot(x2)
plot(y)
```

```python
series ={}
series[0] = x  # the first term will be always the reference signal 
series[1]= x2
series[2]= y
```

```python
taylor_diag.Taylor_diag(series,['x','x2','y'])
```
